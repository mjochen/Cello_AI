# PyCaret

Just one more: PyCaret. Will the model that trains a lot of models outperform our hard work?

We start by importing the data.

```python
import pickle

# Load the pickle file
with open('exports/non_linear_data.pkl', 'rb') as file:
    data_dict = pickle.load(file)

# Display the loaded data
X = data_dict["X"]
y = data_dict["y"]
X_train = data_dict["X_train"]
X_test = data_dict["X_test"]
y_train = data_dict["y_train"]
y_test = data_dict["y_test"]
```

... and converting it to a pandas dataframe with all X-columns (which is just one column, but still) and appending the target-column y.

```python
import pandas as pd

# Convert the data to pandas DataFrames
df = pd.DataFrame(X, columns=[f'feature_{i}' for i in range(X.shape[1])])
df['target'] = y
```

Next we'll import all the regression models PyCaret has and setup an experiment. Let it compare the models to, while we're at it.

```python
from pycaret.regression import *

s = setup(df, target = 'target', session_id = 123)
best = compare_models()
```

The best model has an RMSE of 2.14. And not to brag, but we got to 2.08. Confusing to note that the lightgbm was actually the best model, so in a production environment we'd now go back and train and tune that model.

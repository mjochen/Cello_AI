If you already have a biased dataset, you can assign weights to account for the underrepresentation.

    Give higher statistical weight to leopards (and other under-observed species) based on known or estimated population proportions.

Example:

```python
# In pandas
df['weight'] = df['species'].map({
    'zebra': 1.0,
    'wildebeest': 1.0,
    'antelope': 1.2,
    'giraffe': 1.5,
    'leopard': 10.0  # Heavily upweighted
})
```

Model-Based Correction

Use a propensity score model or Bayesian adjustment to estimate the likelihood of each observation being selected, and correct for that.

    Estimate: P(selected | species, location, time)

    Use inverse probability weighting: give each observation a weight of 1 / P(selected)
